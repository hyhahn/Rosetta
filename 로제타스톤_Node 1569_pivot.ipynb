## Step 1 : Groupby()하여 계산 진행

```python
df2 = df.groupby(['연구분야대분류','사업화형태']).mean()[['지원과제수']]
df2
```

## (임시변통)  컬럼을 새롭게 생성, 멀티인덱스에 해당하는 값

```python
df2.index.get_level_values(0)
```

```python
df2['연구분야대분류'] = df2.index.get_level_values(0)
```

```python
df2['사업화형태'] = df2.index.get_level_values(1)
```

```python
df2
```

## pivot

```python
df2.pivot(index='연구분야대분류', columns='사업화형태', values='지원과제수')
```
