## df[['이전과제번호','과제번호','컬럼3']].apply(lambda x : fun2(x), axis=1)
- axis 조심

```python
def fun1(x):
    
    if x[0] == x[1]:
        return x[0]
    else:
        return None
```

```python
def fun2(x):
    
    if x[0] == x[1]:
        return True
    else:
        return False
```

```python
df['과제번호비교결과'] = df[['이전과제번호','과제번호','컬럼3']].apply(lambda x : fun2(x), axis=1)
```
